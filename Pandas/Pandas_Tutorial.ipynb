# PART 1 : BASICS

**1. Series**

What is a Series?

Answer: Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.).

```python
# variation 1
sample_series = pd.Series(data=[1, 2, 3, 4, 5], index=[10, 11, 12, 13, 14], dtype="int32")
print(sample_series)
```

```python
# variation 2
sample_series = pd.Series(data=[1, 2, 3, 4, 5], index=[20, 21, 22, 23, 24], dtype="float32")
print(sample_series)
```

```python
# variation 3
sample_series = pd.Series(data=[1, 2, 3, 4, 5], index=None, dtype="float32")
print(sample_series)
```

```python
# Variation 4 
animal_size = ["giant", "big", "medium", "small", "tiny"]
animal_name = ["Dinosaur", "Elephant", "Cow", "Cat", "chick"]

# Make the series
animal_series = pd.Series(data=animal_name, index=animal_size, dtype="str", name="animals")
print(animal_series)
```

```python
# adding data using loc and iloc
animal_series.loc["molecular"] = "bacteria"

# iloc possible but not very explainable -- ??
print(animal_series)
```

**LOC VS ILOC**

**loc** : loc gets rows (and/or columns) with particular labels. <br>
**iloc** : iloc gets rows (and/or columns) at integer locations.

```python
# lets use the loc
small_animal = animal_series.loc["small"]
print(f"small animal from animal series : {small_animal}")

# let's use iloc on animal_series
who_animal = animal_series.iloc[1]
print(f"animal at index 1 : {who_animal}")

# lets use the iloc operator
val_at_index_four = sample_series.iloc[4]
print(f"Value at index 4 : {val_at_index_four}")
```

```python
# Method 2 SIMILAR TO LOC
giant_animal = animal_series["giant"]
print(f"giant animal from animal series : {giant_animal}")
```

**VALUE COUNTS**

Value counts can help you to understand the distrubution of various value across the series data

```python
vc_data = pd.Series(data=[1, 2, 3, 3, 4, 4, 4, 5, 5, 5, 5, 5, 8, 8, 9, 9, 9, 9, 9, 9,])
print(vc_data)
print("\n")
print(vc_data.value_counts())
```

# 16. Concat Multiple dataframes <br>

We can concat multiple dataframe together using pd.concat functions. Note if you are concatinating along the row the column names and alignment should be same and if you are concatinating along the columns the number of datapoints should be the same.

```python
# let's make three dataframe and concat them 
a = pd.DataFrame({
    "age" : [10, 11, 12, 13, 14, 15],
    "name" : ["a", "b", "c", "d", "e", "f"]
})

b = pd.DataFrame({
    "marks": [99, 100, 92, 90, 95, 94],
    "subject" : ["maths", "physics", "english", "biology", "chemistry", "politics"]
})

concat_df_along_columns = pd.concat([a, b], axis=1)
concat_df_along_columns
```

```python
c = pd.DataFrame({
    "age" : [20, 21],
    "name" : ["g", "h"],
    "marks": [89, 83],
    "subject": ["cs", "sw"]
})

concat_df_along_rows = pd.concat([concat_df_along_columns, c], axis=0).reset_index(drop=True)
concat_df_along_rows
```

# 17. Performing basic operation on the dataframe

```python
# Mean marks
mean_marks = concat_df_along_rows["marks"].mean()
print(f"mean of columns marks : {mean_marks}")

# Median marks
median_marks = concat_df_along_rows["marks"].median()
print(f"median of columns marks : {median_marks}")

# Max marks
max_marks = concat_df_along_rows["marks"].max()
print(f"median of columns marks : {max_marks}")

# Min marks
min_marks = concat_df_along_rows["marks"].min()
print(f"median of columns marks : {min_marks}")
```
